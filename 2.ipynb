```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("iris (2).csv")
x=data['SepalLengthCm']
y=data['SepalWidthCm']
z=data['PetalWidthCm']
```

```python
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Scatter Plot")
```

```python
sns.boxplot(data=data)
plt.title("Box Plot")
```

```python
select=data.iloc[:,[1,2,3,4]]
sns.heatmap(select.corr(),annot=True)
plt.title("Heat Map")
plt.show()
```

```python
plt.tricontour(x,y,z,levels=20,cmap="jet")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Contour Plot")
```

```python
ax=plt.axes(projection="3d")
ax.set_title("3d")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.plot_trisurf(x,y,z,cmap="jet")
```
