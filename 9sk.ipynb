```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import linkage,dendrogram
```

```python
data=load_iris().data
data=data[:6]
```

```python
def proxi():
    n=data.shape[0]
    pm=np.zeros((n,n))
    for i in range(n):
        for j in range(i+1,n):
            pm[i,j]=pm[j,i]=np.linalg.norm(data[i]-data[j])
    return pm
    
```

```python
print(proxi())
```

```python
def plot(method):
    link=linkage(data,method=method)
    dendrogram(link)
    plt.title(method.capitalize()+"Linkage")
    plt.show()
```

```python
plot("single")
```

```python
plot("complete")
```
