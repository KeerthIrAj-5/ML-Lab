```python
def search(data):
    x=input("Enter element")
    for i in range(len(data)):
        if(data[i]==x):
            print(f"{x} at {i}")
            return
        
    print("Not found")
def sort(data):
    for i in range(len(data)):
        for j in range(len(data)-1-i):
            if(data[j]>data[j+1]):
                data[j],data[j+1]=data[j+1],data[j]
def update(data,key):
    if key not in range(len(data)):
        print("Key not found")
    else:
        x=input("Enter new element")
        data[key]=x
list1=[]
def listop():
    print("List operations")
    print("1.Insert\n2.search\n3.update\n4.delete\n5.display\n6.sort\n7.menu")
    while True:
        ch=int(input("Enter choice"))
        if ch==1:
            x=input("Enter element")
            list1.append(x)
        elif ch==2:
            search(list1)
        elif ch==3:
            key=int(input("Enter key"))
            update(list1,key)
        elif ch==4:
            try:
                x=input("Enter element")
                list1.remove(x)
            except:
                print("Not exists")
        elif ch==5:
            print(list1)
        elif ch==6:
            sort(list1)
        elif ch==7:
            return
        else:
            print("Invalid choice")

tuple1=tuple()
def tupleop():
    tuple1=tuple(i for i in input().split())
    print("Tuple operations")
    while True:
        print("1.Search\n.2.Display\n3.Return")
        ch=int(input("Enter choice"))
        if ch==1:
            x=input("enter element")
            if x in tuple1:
                print("present")
            else:
                print("Absent")
        elif ch==2:
            print(tuple1)
        elif ch==3:
            return
        else:
            print("Invalid choice")
set1=set()
def setop():
    print("Set operations")
    print("1.Insert\n2.search\n4.delete\n5.display\n7.menu")
    while True:
        ch=int(input("Enter choice"))
        if ch==1:
            x=input("Enter element")
            set1.add(x)
        elif ch==2:
            x=input("Enter element")
            if x in set1:
                print("Present")
            else:
                print("Absent")
        elif ch==4:
            try:
                x=input("Enter element")
                set1.remove(x)
            except:
                print("Not exists")
        elif ch==5:
            print(set1)
        elif ch==7:
            return
        else:
            print("Invalid choice")
dict1=dict()
def dictop():
    print("Dict operations")
    print("1.Insert\n2.search\n3.update\n4.delete\n5.display\n7.menu")
    while True:
        ch=int(input("Enter choice"))
        if ch==1:
            key=input("Enter key")
            x=input("Enter element")
            dict1[key]=x;
        elif ch==2:
            x=input("Enter element")
            for key,value in dict1.items():
                if(value==x):
                    print(key)
        elif ch==3:
            key=(input("Enter key"))
            if key in dict1:
                x=input("Enter new element")
                dict1[key]=x
            else:
                print("Not present")
        elif ch==4:
            try:
                x=input("Enter element")
                del dict1[x]
            except:
                print("Not exists")
        elif ch==5:
            print(dict1)
        elif ch==7:
            return
        else:
            print("Invalid choice")

while True:
    print("1.List\n2.Tuple\n3.Set\n4.Dictionary\n5.Exit")
    ch=int(input("Enter choice"))
    if ch==1:
        listop()
    elif ch==2:
        tupleop()
    elif ch==3:
        setop()
    elif ch==4:
        dictop()
    elif ch==5:
        break
    else:
        print("Invalid choice")
    
```
