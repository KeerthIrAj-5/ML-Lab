```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier as KNN
```

```python
glass=pd.read_csv("glass (1).csv")
fruit=pd.read_csv("fruits (1).csv")
datasets=["Glass","Fruits"]
xg=glass.drop("Type",axis=1).values
yg=glass["Type"].values
xf=fruit[["mass","width","height","color_score"]].values
yf=fruit["fruit_label"].values
```

```python
k_values=[3,5,7]
test_split=[0.1,0.3]
distances=["manhattan","euclidean"]
```

```python
for data in datasets:
    print(f"dataset:{data}")
    for k in k_values:
        for p,distance in enumerate(distances):
            for ts in test_split:
                print(f"K:{k}Distance:{distance}")
                print(f"Split:{ts}")
                if data=="Glass":
                    x_train,x_test,y_train,y_test=train_test_split(xg,yg,test_size=ts,random_state=42)
                else:
                    x_train,x_test,y_train,y_test=train_test_split(xf,yf,test_size=ts,random_state=42)
                knn=KNN(n_neighbors=k,p=p+1,weights="distance")
                knn.fit(x_train,y_train)
                accuracy=knn.score(x_test,y_test)
                print(f"Accuracy:{accuracy:.2}")
                print("------------")
            print("-----------------------")
        print("--------------------------------")
```
