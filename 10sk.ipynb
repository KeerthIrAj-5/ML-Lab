```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
x=load_iris().data
y=load_iris().target
print("Shape is ",x.shape)
plt.scatter(x[:,0],x[:,1],c=y,cmap="jet")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Original")
plt.show()
```

```python
pca=PCA(n_components=2)
pca.fit(x)
xp=pca.transform(x)
print("Shape is ",xp.shape)
plt.scatter(xp[:,0],xp[:,1],c=y,cmap="jet")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA")
plt.show()
```

```python
lda=LDA(n_components=2)
lda.fit(x,y)
xl=lda.transform(x)
print("Shape is ",xl.shape)
plt.scatter(xl[:,0],xl[:,1],c=y,cmap="jet")
plt.xlabel("LD1")
plt.ylabel("LD22")
plt.title("LDA")
plt.show()
```
